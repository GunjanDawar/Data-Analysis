## Q2) Which toss decision is best for each ground?

```python
# ipl_df['venue'].value_counts()
```

```python
team1=ipl_df.query("team1==winner")
team2=ipl_df.query("team2==winner")
g_data=team1['venue'].value_counts().index
# g_data

for ground in g_data:
    if(team1['venue'].value_counts()[ground] > team2['venue'].value_counts()[ground]):
        print(ground +  " : is best for  Batting")
    else:
        print(ground + " : is best for Fielding")
                                 
    
```
